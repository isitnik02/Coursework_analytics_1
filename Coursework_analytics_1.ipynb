```python
# импорт библиотек для работы с данными
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import datetime
import missingno as msno
import itertools
```

```python
# объявление переменных, загрузка данных в датасеты

main = pd.DataFrame()
df_courses = pd.read_csv('courses.csv', index_col=0)
df_course_contents = pd.read_csv('course_contents.csv')
df_progresses = pd.read_csv('progresses.csv') 
df_progress_phases = pd.read_csv('progress_phases.csv')

# создание отдельного датасета для задачи расчета среднего возраста студентов курсов  

df_students = pd.read_csv('students.csv')

# создание основного (общего) датасета

main = df_progresses.merge(df_courses, left_on='course_id', right_on='id')

# переименование столбцов для удобства дальнейшей работы с данными

main.rename(columns={'id_x': 'progress_id', 'title': 'course_title', 'field': 'course_field'},
            inplace=True)

# удаление столбца 'id_y', т.к. в датасете уже есть столбец course_id с аналогичными данными

main.drop(['id_y'], axis=1, inplace=True)

# обогащение датасета main доп.даннми о курсах и об прогрессе прохождения курсов

main = main.merge(df_course_contents, left_on='course_id', right_on='course_id')
main = main.merge(df_progress_phases,
                        left_on=['progress_id', 'module_number', 'lesson_number'], 
                        right_on=['progress_id', 'module_number', 'lesson_number'])

# преобразование данных в колонках 'start_date' и 'finish_date' из типа object в тип даты

main['start_date'] = main['start_date'].astype('datetime64[ms]')
main['finish_date'] = main['finish_date'].astype('datetime64[ms]')
```

```python
# для определения минимального, максимального, среднего и медианного возраста студентов
# потребуется очистить датасет с данними по студентам от пустых значений по колонке "birthday"

df_students = df_students[df_students['birthday'].notnull()].copy()

# проверка результата. Датасет очищен от пропущенных значений по дате рождения, при этом сохранены
# записи по студентам, у которых заполнен возраст и не заполнено поле город

df_students.isnull().sum()
```

```python
# для выполнения задач по определению возрастов необходимо добавить колонку "возраст" ('age')
# в описании задания не сказано от какой даты расчитать возраст, поэтому будет взята макс.финальная
# дата окончания

df_students['age'] = df_students['birthday'].apply(
    lambda x: main['finish_date'].max().year - datetime.date.fromisoformat(x).year)
```

```python
# обзор датасета. выявлен дубликат 'id_'

df_students.info()
```

```python
# удаление дублирующейся колонки

df_students.drop('id_', axis=1, inplace=True)
```

```python
# анализ полученных данных показал, что существуют некорректные данные по возрасту
# принято решение очистить датасет от студентов с возрастом младше 18 лет и не старше 70 лет
# безусловно, несовершеннолетние в возрасте от 14 лет могут сами заключать договора на получение
# услуг в сфере образования, с согл.родителей, но я решил опустить данный момент и установить 
# ограничение минимум 18 лет.

df_students['age'].describe()
```

```python
# фильтр датасета df_students по возрасту

df_students = df_students.query('age >= 18 and age <= 60')

# итогцц задачи "посчитать минимальный, максимальный, средний, медианный возраст студентов"

df_students['age'].describe()
```

```python
# выполнение послдней подзадачи из пункта №1
# "посчитать минимальный, максимальный, средний, медианный возраст студентов на каждом курсе"

main.merge(df_students, left_on='student_id', right_on='id')\
    .groupby('course_title')['age'].agg(['min', 'max', 'mean', 'median'])
```

Задача № 2: Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы

```python
# построение столбчатой диаграммы (bar-chart), отражающий количество студентов на каждом курсе. 

# установил размеры графика
fig, ax = plt.subplots(figsize=(12, 8))

# построение граифка 
main.pivot_table(values='student_id', index='course_title', aggfunc=lambda x: x.unique().size)\
    .sort_values('student_id', ascending=False)\
    .plot.bar(ax=ax, legend=False, color='g')

# подпись тиков
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.set(xlabel='', ylabel='')

# вывод графика в блокнот
plt.show()
```

Задача № 3: Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

```python
# построение графика
fig, ax = plt.subplots(figsize=(12, 8))

# создание временного датасета с курсами и студентами
df = main.pivot_table(values='student_id', index='course_title', columns='course_field', aggfunc=lambda x: x.unique()\
            .size, margins=True).sort_values('All', ascending=True)\
            .drop(index=['All'], columns=['All'])

# Построение диаграммы, отражающей количество студентов на курсах
df.plot.barh(ax=ax, title='Количество студентов на курсах', stacked=True, alpha=0.1)

# Построение линии медианы на диаграмме
df = df.stack().reset_index()
ax.vlines(df[0].median(), ax.get_ylim()[0], ax.get_ylim()[1], color='r', linestyles='dashed', linewidth=1)

# Скрытие контура диаграммы
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Вывод названия осей
ax.set(xlabel='', ylabel='')

# Вывод диаграммы на страницу
plt.show()
```

```python
# создание датасета по прогрессу выполнения домашних работ
df_progress = main.loc[main['is_homework']==True, ['course_title', 'status', 'start_date', 'finish_date']].copy()

# преобразование формата дат к краткому виду
df_progress['start_date'] = df_progress['start_date'].astype('datetime64[M]')
df_progress['finish_date'] = df_progress['finish_date'].astype('datetime64[M]') + pd.tseries.offsets.MonthEnd(0)

# указание дат окончания для несданных и незавершенных работ
df_progress.loc[df_progress['status']=='fail', 'finish_date'] = df_progress['start_date'] + pd.tseries.offsets.MonthEnd(0)
df_progress.loc[df_progress['status']=='start', 'finish_date'] = df_progress['finish_date'].max()
df_progress.loc[df_progress['status']=='wait', 'finish_date'] = df_progress['finish_date'].max()

# создание списка месяцев
column_list = pd.date_range(df_progress['start_date'].min(), df_progress['finish_date'].max(), freq='M')

# расширение датасета путем добавления столбцов из списка месяцев и формирование матрицы соответствия даты и месяца
for item in column_list:
    df_progress.loc[(df_progress['start_date']<=item) & (item<=df_progress['finish_date']), item] = 1

# формирование датасета с количеством выполненных домашних работ в каждый месяц
df_progress = df_progress.groupby('course_title').sum().stack().reset_index()\
              .rename(columns={'level_1': 'homework_date', 0: 'homework_count'})

for title in df_progress['course_title'].unique(): # определение прогресса выполнения домашних работ за каждый месяц
    indexes = df_progress[df_progress['course_title']==title].index
    for index in indexes:
        df_progress.loc[index, 'homework_progress'] = df_progress.loc[indexes[0]:index, 'homework_count'].sum()
        
df_progress.query('homework_progress > 0').head(5) # просмотр первых 5-ти позиций
```

Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле)

```python
# формирование нового датасета с определением дат начала и окончания всех домашних заданий в модуле
df_conversion = main.loc[main['is_homework']==True, 
                        ['course_title', 'course_field', 'module_number', 'module_title', 'student_id', 'start_date', 'finish_date']]\
                    .groupby(['course_title', 'course_field', 'module_number', 'module_title', 'student_id'])\
                    .agg({'start_date': np.min, 'finish_date': np.max})

# очистка датасета от пустых значений по датам старта и окончания, подсчет количества студентов начавших и завершивших обучение 
df_conversion = df_conversion.assign(student_start=df_conversion['start_date'].notnull(),
                                     student_finish=df_conversion['finish_date'].notnull())\
                  .groupby(['course_title', 'course_field', 'module_number', 'module_title'])\
                  .agg({'student_start': np.sum, 'student_finish': np.sum, 'start_date': np.min, 'finish_date': np.max})\
                  .sort_index().reset_index()

# преобразование дат к краткому виду
df_conversion['start_date'] = df_conversion['start_date'].astype('datetime64[D]')
df_conversion['finish_date'] = df_conversion['finish_date'].astype('datetime64[D]')

# расчет конверсии и добавление значений в датасет
for course in df_conversion['course_title'].unique():
    for index, module_number in df_conversion.loc[df_conversion['course_title']==course, 'module_number'].items():
        
        # подсчет количества студентов, начавших обучение в текущем модуле
        student_start = df_conversion.loc[index, 'student_start']
        
        # подсчет количества студентов, завершивших обучение в предыдущем модуле
        student_finish = df_conversion.loc[(df_conversion['course_title']==course) & 
                        (df_conversion['module_number'] < module_number), 'student_finish'].tail(1).min()
        
        # подсчет конверсии текущего модуля
        df_conversion.loc[index, 'conversion'] = student_start / student_finish if student_finish > 0 else 0

```
